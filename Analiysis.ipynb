```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('results.csv', sep = ';')
```

```python
data.head()
```

# Best values

```python
data.sort_values('dif_mean', ascending=False).head(15)
```

Widać, że najlepsze wyniki otrzymujemy dla where_w = 1 i const_2 = 1 

```python
keys = ['sigma', 'const_1', 'const_2', 'where_w']
for key in keys:
    print('Mean for {}'.format(key))
    print(data.groupby(key).mean()['dif_mean'])
    print()
    print('STD for {}'. format(key))
    print(data.groupby(key).std()['dif_mean'])
    print('*'* 80)
```

Where_w najprostrze w interpretacji, zmienia najwięcej, powinno być zawsze na 1. Bardzo stabilne.

const_2 w najlepszych wynikach wydaje się być ustalone na 1.0. Ale najlepsze i najbardziej stabilne wyniki daje dla 0.8. Widocznie optimum z nim zwiazane powinno być na innych parametrach sigma in const_1

sigma i const_1 wydają się być mniej wiecej losowe, const_1 powinno być duże. 

# Jakie sigma i const_1

```python
dd = data[(data.where_w == 1) & (data.const_2 == 1.0)][['sigma', 'const_1', 'dif_mean']]
dd.plot(kind= 'scatter', x = 'sigma', y = 'const_1', c = 'dif_mean')
plt.show()
```

Sigma powinno być małe
